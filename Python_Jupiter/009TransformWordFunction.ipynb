```python
def Transform(sth):
    translation = ' '
    for i in sth :
        if i.lower() in 'áàảãạăắằẳẵặâấầẩẫậ':
            if i.isupper():
                translation = translation + 'A'
            else :
                translation = translation + 'a'
        else:
            translation = translation + i 

    return translation
```

```python
print(Transform('Sáng'))
print(Transform('SÁNG'))
```
