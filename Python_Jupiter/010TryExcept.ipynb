```python
try:
    num1 = 'hai'
    num2 = 0

    num3 = num1 / num2 

    print(num3)
    
except ZeroDivisionError:
    print('Mẫu số phải khác 0')
except TypeError:
    print('Không thể thực hiện phép chia giữa chữ và số!')
except Exception as e:
    print(f'Lỗi hệ thống: {e}')
```
