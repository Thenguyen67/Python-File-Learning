```python
str1 = "tôi đang học lập trình"
print(str1.upper()) #Chuyển toàn bộ kí tự từ thường sang hoa
```

```python
str1 = "TÔI ĐANG HỌC LẬP TRÌNH"
print(str1.lower()) # Chuyẻn toàn bộ kí tự từ hoa sang thường
```

```python
str1 = "a"
str2 = "A"

print(str1.islower()) #Kiểm tra liệu chuỗi có phải chỉ có kí tự in thường
print(str2.isupper()) #Kiểm tra liệu chuỗi có phải chỉ có kí tự in hoa
print(str1.upper().isupper())
```

```python
str1 = "tôi đã"
print(len(str1)) #In ra số lượng kí tự trong chuỗi bao gồm cả khoảng trắng
print(str1[0]) #In ra kí tự (Không bao gồm khoảng trắng)
print(str1[1])
print(str1[2])
```

```python
str1 = "tôi đang học lập trình"
print(str1.index("h")) #Tìm vị trí xuất hiện lần đầu của kí tự 
print(str1.index("đang"))
```

```python
str1 = "tôi đang học lập trình"
print(str1.replace("đang", "chưa")) #Thay thế kí tự trong chuỗi thành kí tự mới
```
